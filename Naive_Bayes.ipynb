```python
import pandas as pd
import numpy as np
df = pd.read_csv("titanic.csv")
df.head()
```

```python
#survivedに影響しないcolumnを予めremove
df.drop(["PassengerId","Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis="columns", inplace =True)
df.head()
#inplace=Trueによって、本体dataを直接変えるから、inplace=Trueなしでdf.head()やっても、何も変化は見られない。
```

```python
# Replace zeroes

# Replace zeroes in 'Age' column with NaN
df['Age'] = df['Age'].replace(0, np.NaN)
# Compute the mean of 'Age' column, skipping NaN values
mean_age = int(df['Age'].mean(skipna=True))
# Replace NaN values in 'Age' column with the computed mean
df['Age'] = df['Age'].replace(np.NaN, mean_age)

#NaNがあるとGaussianNBはerrorになる→実際、エラーになった
#以下のコードにてどのcolumnがNaN持っているかは確認可能
#columns_with_nan = df.columns[df.isnull().any()].tolist()
#print(columns_with_nan)
#どうやらfare columnにもNaNがあったらしいので、removeする。

# Compute the mean of 'Age' column, skipping NaN values
mean_fare = int(df['Fare'].mean(skipna=True))
# Replace NaN values in 'Age' column with the computed mean
df['Fare'] = df['Fare'].replace(np.NaN, mean_fare)
```

```python
#Sex columnにおけるmale/femaleではcomputerが理解できないからnumberに直す、これやらないとモデルにfitした時にerror起こる
#A machine learning model doesn't understand these words, so we need to convert them into numbers.

df = pd.get_dummies(df)
df.head()
```

```python
# split dataset
from sklearn.model_selection import train_test_split
inputs = df.iloc[:, 1:]
target = df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(inputs, target, random_state=42, test_size=0.15)
```

```python
#確認
print(len(y_train))
print(len(y_test))

#スマートに書く
#for takeisika in [X_train, X_test, y_train, y_test]: print(len(takeisika))

print (y_test.head (10))
print (y_train.head (10))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)

#正解y_testは下記
print(y_test.head(15))
```

```python
# Evaluate Model
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print (cm)

from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```
